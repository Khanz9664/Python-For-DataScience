#
## 1. Basics of Matplotlib
### 1.1 Importing Matplotlib and Plotting a Simple Line Graph

```python
import matplotlib.pyplot as plt
```

#
### Simple line plot

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label="y = 2x")# Plots the data points defined by x and y on a 2D graph. The label argument sets the legend entry for the line.
plt.title("Simple Line Graph")# Sets the title of the graph.
plt.xlabel("X-axis")# Labels the x-axis.
plt.ylabel("Y-axis")# Labels the y-axis.
plt.legend()# Displays a legend for the graph, using the label defined in plt.plot.
plt.show()# Renders and displays the graph.
```

#
## 2. Customizing Plots
### 2.1 Adding Styles and Colors

```python
# Using styles
plt.style.use('seaborn-darkgrid')

# Customizing line styles and colors
plt.plot(x, y, color='blue', linestyle='--', marker='o', label="y = 2x")
plt.title("Styled Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

#
### 2.2 Adding Annotations

```python
plt.plot(x, y, marker='o')
plt.title("Line Graph with Annotation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Annotating a specific point
plt.annotate('Peak Point', xy=(5, 10), xytext=(4, 9),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

#
## 3. Multiple Plots and Subplots
### 3.1 Multiple Plots in a Single Figure

```python
x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 4, 6, 8, 10]

plt.plot(x1, y1, label="y = x^2")
plt.plot(x2, y2, label="y = 2x")
plt.title("Multiple Lines in One Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

#
### 3.2 Creating Subplots

```python
# Creating subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# First subplot
axs[0].plot(x1, y1, color='red', label='y = x^2')
axs[0].set_title("First Subplot")
axs[0].legend()

# Second subplot
axs[1].plot(x2, y2, color='blue', label='y = 2x')
axs[1].set_title("Second Subplot")
axs[1].legend()

plt.tight_layout()
plt.show()
```

#
## 4. Advanced Plot Types
### 4.1 Bar Charts

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]

plt.bar(categories, values, color='green')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#
### 4.2 Scatter Plots

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 500 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title("Scatter Plot")
plt.show()
```

#
## 5. 3D Plotting

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

## 7. Animations

```python
#This will enable interactive plots and allow the animation to display in the notebook.
%matplotlib notebook
```

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x, y = [], []

line, = ax.plot([], [], 'r-')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    return line,

def update(frame):
    x.append(frame)
    y.append(frame**2)
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=range(10), init_func=init, blit=True)
plt.show()
```

## 12. Working with Dates

```python
import matplotlib.dates as mdates
import pandas as pd
```

```python
# Generate a time series
dates = pd.date_range("2023-01-01", periods=10)
values = np.random.randint(1, 10, size=10)

plt.figure(figsize=(10, 5))
plt.plot(dates, values, marker='o')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.title("Time Series Data")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.show()
```

## 15. Animations (Advanced)
### 15.1.Animated Scatter Plot

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()
x, y = [], []
sc = ax.scatter([], [], s=100)

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    return sc,

def update(frame):
    x.append(np.random.rand())
    y.append(np.random.rand())
    sc.set_offsets(np.c_[x, y])
    return sc,

ani = animation.FuncAnimation(fig, update, frames=20, init_func=init, blit=True)
plt.show()
```

#
## 16.Multi-Axis and Twin Axes
### 16.1 Secondary Y-axis

```python
fig, ax1 = plt.subplots()

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax1.plot(x, y1, 'b-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Sin', color='blue')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'r--')
ax2.set_ylabel('Cos', color='red')

plt.title("Twin Axes Example")
plt.show()
```

## 17.3D Plotting (Advanced)
### 17.1 3D Wireframe Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_wireframe(X, Y, Z, color='green')
plt.title("3D Wireframe")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def plot_3d(angle):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.view_init(elev=30, azim=angle)
    ax.plot_wireframe(X, Y, Z)
    plt.show()

interact(plot_3d, angle=(0, 360, 10))
```
